**PREPARING DATA**

```python
iris_dataset=iris_dataset.drop(columns=['Id'])
```

```python
iris_dataset.head()
```

**CHANGING DATAS INTO NUMERICAL**

```python
iris_dataset['Species'].replace({"Iris-setosa":1,"Iris-versicolor":2,"Iris-virginica":3},inplace=True)
iris_dataset
```

**TO CREATE ARRAY FOR TRAINING AND TESTING**

```python
a=pd.DataFrame(iris_dataset,columns=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]).values
```

```python
a
```

```python
b=iris_dataset.Species.values.reshape(-1,1)
b
```

**TRAINING A MODEL**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
a_train,a_test,b_train,b_test=train_test_split(a,b,test_size=0.2)
```

```python
model=LogisticRegression()
model.fit(a_train,b_train)
```

**PREDICTIONS**

```python
#To check accuracy
prediction=model.predict(a_test)
print(accuracy_score(b_test,prediction)*100)
```

```python
print(classification_report(b_test,prediction))
 #IN OUTPUT
 #1=IRIS-SETOSA
 #2=IRIS-VERSICOLOR
 #3=IRIS-VIRGINICA
```

```python
new=np.array([[5.3,2.5,4.6,1.9],[4.4,2.9,1.4,0.2]])
#prediction from an array
pre=model.predict(new)
print("Prediction is : {}".format(pre))
 #IN OUTPUT
 #1=IRIS-SETOSA
 #2=IRIS-VERSICOLOR
 #3=IRIS-VIRGINICA
```
